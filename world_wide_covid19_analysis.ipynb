# Importing libraris

```python
import pandas as pd
import numpy as np
from pivottablejs import pivot_ui
import qgrid
!jupyter nbextension enable --py --sys-prefix qgrid
!jupyter nbextension enable --py --sys-prefix widgetsnbextension
```

# Confirmed cases analysis

```python
confirmed=pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
confirmed['total']=confirmed.iloc[:,4:].sum(axis=1)
confirmed_1=confirmed[['Province/State','Country/Region','Lat','Long','total']]
confirmed_pivot=pd.pivot_table(confirmed_1,index=["Country/Region"],values=["total"],aggfunc=["sum"])
confirmed_pivot.columns=confirmed_pivot.columns.droplevel(level=0)
confirmed_pivot.reset_index(inplace=True)
confirmed_pivot.nlargest(10,'total').plot(x=0,kind='bar',figsize=(10,10),title="Top 10 countries affected by covid 19",)
```

# Confirmed deaths Analysis

```python
confirmed_deaths=pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
confirmed_deaths['total']=confirmed_deaths.iloc[:,4:].sum(axis=1)
confirmed_deaths_1=confirmed_deaths[['Province/State','Country/Region','Lat','Long','total']]
confirmed_deaths_pivot=pd.pivot_table(confirmed_deaths_1,index=["Country/Region"],values=["total"],aggfunc=["sum"])
confirmed_deaths_pivot.columns=confirmed_deaths_pivot.columns.droplevel(level=0)
confirmed_deaths_pivot.reset_index(inplace=True)
confirmed_deaths_pivot.nlargest(10,'total').plot(x=0,kind='bar',figsize=(10,10),title="Top 10 countries with highest deaths due to  covid 19")
```

# Confirmed Recovered Analysis

```python
confirmed_recovered=pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")
confirmed_recovered['total']=confirmed_recovered.iloc[:,4:].sum(axis=1)
confirmed_recovered_1=confirmed_recovered[['Province/State','Country/Region','Lat','Long','total']]
confirmed_recovered_pivot=pd.pivot_table(confirmed_recovered_1,index=["Country/Region"],values=["total"],aggfunc=["sum"])
confirmed_recovered_pivot.columns=confirmed_recovered_pivot.columns.droplevel(level=0)
confirmed_recovered_pivot.reset_index(inplace=True)
confirmed_recovered_pivot.nlargest(10,'total').plot(x=0,kind='bar',figsize=(10,10),title="Top 10 countries recovered count from covid 19")
```
